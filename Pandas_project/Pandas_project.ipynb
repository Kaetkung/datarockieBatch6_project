```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
print(sample[sample['Order Date'].dt.year == 2017]['Sales'].describe())
print(sample[sample['Order Date'].dt.year == 2017]['Sales'].sum())

print("Total sales = 484247.4981, avg = 242.97, SD = 754.05")
```

```python
# TODO 06 - which Segment has the highest profit in 2018
print(sample[sample['Order Date'].dt.year == 2018].groupby('Segment')['Profit'].sum().sort_values(ascending=False))

print("The highest profit in 2018 is Consumer segment = 28460.1665")
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
sample[(sample['Order Date'] >= '2019-04-15')&(sample['Order Date'] <= '2019-12-31')].groupby('State')['Sales'].sum()\
    .sort_values(ascending=True).head(5)
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25%
sample[sample['Order Date'].dt.year==2019].query("Region=='West'|Region=='Central'")['Sales'].sum()*100\
    /sample[sample['Order Date'].dt.year==2019]['Sales'].sum()
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
topProduct = sample[(sample['Order Date'].dt.year >= 2019) & (sample['Order Date'].dt.year <= 2020)].value_counts('Product Name')\
    .sort_values(ascending=False).sort_values(ascending=False).head(10).reset_index()
#renaem header
topProduct.columns = ['Top 10 Product', 'Number of Orders'] 

topProduct
```

```python
#top 10 popular products in terms of total sales during 2019-2020
topProduct_bysales = sample[(sample['Order Date'].dt.year >= 2019) & (sample['Order Date'].dt.year <= 2020)]\
    .groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10).reset_index()

topProduct_bysales.columns = ['Top 10 Product','Total sales']
topProduct_bysales
```

```python
OrdersVsSales = pd.concat([topProduct, topProduct_bysales], axis=1)
OrdersVsSales
```
